Q.3:Provide two real-world scenarios where removing stopwords can be harmful.

Ans:
Removing stopwords is not always good. In some real-world cases, it can change the meaning of a sentence. For example, in sentiment analysis, a sentence like “I do not like this product” becomes “I like product” if stopwords like “not” are removed, which gives the wrong (positive) meaning. Another case is in question answering or search systems, where stopwords help understand intent. For example, “What is the best phone?” vs “best phone”—removing words like “what” and “is” can make the system lose the actual question context. So, removing stopwords can sometimes harm understanding instead of helping.

Q.4:Explain the difference between stemming and lemmatization

Ans:

Stemming and lemmatization are both techniques used in NLP to reduce words to their base form, but they work in different ways. **Stemming** simply cuts off word endings using basic rules, without considering the actual meaning, so it may produce words that are not real (for example, *“studies” → “studi”*). In contrast, **lemmatization** uses grammar and context to convert a word into its correct dictionary form, so the output is always meaningful (for example, *“studies” → “study”*). In short, stemming is faster but less accurate, while lemmatization is slower but gives better and more meaningful results.

```python
import re

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    words = text.split()
    words = [w for w in words if len(w) > 2 or w in ['no', 'not']]

    return ' '.join(words)


text = "WOW!!! This is soooo goooood!!! Visit http://example.com now, I have 2 dogs. Email me at test@gmail.com"
print(preprocess_text(text))
```

```python
import re

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    words = text.split()
    words = [w for w in words if len(w) > 2 or w in ['no', 'not']]

    return words


texts = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]


# Analytics
for i, text in enumerate(texts, 1):
    tokens = preprocess_text(text)

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(w) for w in tokens) / total_tokens if total_tokens > 0 else 0

    print(f"{i}. {text}")
    print(f"   Tokens: {tokens}")
    print(f"   Total Tokens: {total_tokens}")
    print(f"   Unique Tokens: {unique_tokens}")
    print(f"   Avg Token Length: {avg_length:.2f}")
    print("-" * 60)
```

**Task 5: Frequency Analysis**

```python
import re
from collections import Counter

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    words = text.split()
    words = [w for w in words if len(w) > 2 or w in ['no', 'not']]

    return words


texts = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

all_tokens = []
for text in texts:
    all_tokens.extend(preprocess_text(text))

# Count frequency
freq = Counter(all_tokens)

# Top 10 most frequent words
print("Top 10 Most Frequent Words:")
for word, count in freq.most_common(10):
    print(word, ":", count)

# Top 5 least frequent words
print("\nTop 5 Least Frequent Words:")
least_words = sorted(freq.items(), key=lambda x: x[1])
for word, count in least_words[:5]:
    print(word, ":", count)
```

**Task 6: Build Full Pipeline**

```python
import re

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    words = text.split()
    words = [w for w in words if len(w) > 2 or w in ['no', 'not']]

    return words, ' '.join(words)


def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean_text = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean_text)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


# Example test
texts = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "I am not happy with this"
]

result = full_pipeline(texts)
print(result)
```

**Task 7: Error Handling**

```python
import re

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""

    text = text.lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    if text == "":
        return [], ""

    words = text.split()
    words = [w for w in words if len(w) > 2 or w in ['no', 'not']]

    return words, ' '.join(words)


def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean_text = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean_text)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


# Test cases (Error Handling)
texts = [
    "",                 # empty string
    "😂😂😂🔥🔥",        # only emojis
    "1234567890",       # only numbers
    "Normal text here"
]

result = full_pipeline(texts)
print(result)
```
